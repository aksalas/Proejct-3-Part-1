```python
import pandas as pd
import numpy as np
```

```python
import os
os.makedirs('Data/',exist_ok=True) # Confirm folder created
os.listdir("Data/")
```

```python
title_aka = pd.read_csv('Data/title-akas-us-only.csv')
title_aka.head()
```

```python
title_basics = pd.read_csv('Data/title.basics.tsv.gz', sep='\t', low_memory=False)
title_basics.head()
```

```python
title_ratings = pd.read_csv('Data/title.ratings.tsv.gz', sep='\t', low_memory=False)
title_ratings.head()
```

```python
title_aka = title_aka.replace({'\\N':np.nan})
title_aka.head()
```

```python
title_basics = title_basics.replace({'\\N':np.nan})
title_basics.head()
```

```python
title_ratings = title_ratings.replace({'\\N':np.nan})
title_ratings.head()
```

```python
title_basics = title_basics.dropna(subset=['genres'], inplace =True)
title_basics = title_basics.dropna(subset=['runtimeMinutes'], inplace =True)
```

```python
title_basics['startYear'] = title_basics['startYear'].astype(float)
title_basics['startYear'].dtype
```

```python
title_basics.to_csv('Data/title.basics.tsv.gz',compression='gzip',index=False)
title_ratings.to_csv('Data/title.ratings.tsv.gz',compression='gzip',index=False)
```
