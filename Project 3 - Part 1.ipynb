```python
import pandas as pd
import numpy as np
```

```python
import os
os.makedirs('Data/',exist_ok=True) # Confirm folder created
os.listdir("Data/")
```

```python
title_aka = pd.read_csv('C:\\Users\\arvin\\Documents\\0626 Data enrichment\\Proejct-3-Part-1\\Data\\title-akas-us-only.csv')
title_aka.head()
```

```python
title_basics = pd.read_csv('https://datasets.imdbws.com/title.basics.tsv.gz', sep='\t', low_memory=False)
title_basics.head()
```

```python
title_ratings = pd.read_csv('https://datasets.imdbws.com/title.ratings.tsv.gz', sep='\t', low_memory=False)
title_ratings.head()
```

```python
title_aka = title_aka.replace({'\\N':np.nan})
title_aka.head()
```

```python
title_basics = title_basics.replace({'\\N':np.nan})
title_basics.head()
```

```python
title_ratings = title_ratings.replace({'\\N':np.nan})
title_ratings.head()
```

```python
title_basics.info()
```

```python
title_basics = title_basics.dropna(subset=['genres'])
title_basics = title_basics.dropna(subset=['runtimeMinutes'])
```

```python
title_basics['startYear'] = title_basics['startYear'].astype(float)
title_basics['startYear'].dtype
```

```python
title_basics['titleType'].value_counts()
```

```python
basics_movie = 'movie'
title_basics = title_basics[title_basics['titleType'] == basics_movie]
title_basics.head()
```

```python
title_basics = title_basics[(title_basics['startYear'] >= 2000) & (title_basics['startYear'] <= 2021)]
title_basics.head()
```

```python
is_documentary = title_basics['genres'].str.contains('documentary',case=False)
title_basics = title_basics[~is_documentary]
```

```python
keepers =title_basics['tconst'].isin(title_aka['titleId'])
keepers
```

```python
title_basics = title_basics[keepers]
title_basics
```

```python
title_basics.to_csv('\\Users\\arvin\\Documents\\0626 Data enrichment\\Proejct-3-Part-1\\Data\\title.basics.csv.gz',compression='gzip',index=False)
title_ratings.to_csv('\\Users\\arvin\\Documents\\0626 Data enrichment\\Proejct-3-Part-1\\Data\\title.ratings.csv.gz',compression='gzip',index=False)
```
